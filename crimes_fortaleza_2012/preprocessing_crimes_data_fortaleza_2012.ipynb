```python
df = homicidios
#df['2012-01-01 '].append(df['Column 2']).reset_index(drop=True)
df = df['2012-01-01'].append(df['2012-02-01']).append(df['2012-03-01']).append(df['2012-04-01']).append(df['2012-05-01']).append(df['2012-06-01']).append(df['2012-07-01']).append(df['2012-08-01']).append(df['2012-09-01']).append(df['2012-10-01']).append(df['2012-11-01']).append(df['2012-12-01']).reset_index()
df = df.rename(columns={0: "Homicidios"})
df
```

```python
months_dict = {1:'2012-01-01',
               2:'2012-02-01',
               3:'2012-03-01',
               4:'2012-04-01',
               5:'2012-05-01',
               6:'2012-06-01',
               7:'2012-07-01',
               8:'2012-08-01',
               9:'2012-09-01',
               10:'2012-10-01',
               11:'2012-11-01',
               12:'2012-12-01'}
months_dict
```

```python
months = []
key = 0
for i in range (len(roubos)*len(roubos.columns)):    
    if (i%len(roubos) == 0):
        key += 1
        #print('key = ', key, end = ' ')
        #print('i = ',i)
    string = months_dict[key]
    months.append(string)
    
len(months)    
```

```python
df['MÃªs'] = months
homicidios2 = df[['Bairro','MÃªs','Homicidios']]
homicidios2
```

```python
df = roubos
#df['2012-01-01 '].append(df['Column 2']).reset_index(drop=True)
df = df['2012-01-01'].append(df['2012-02-01']).append(df['2012-03-01']).append(df['2012-04-01']).append(df['2012-05-01']).append(df['2012-06-01']).append(df['2012-07-01']).append(df['2012-08-01']).append(df['2012-09-01']).append(df['2012-10-01']).append(df['2012-11-01']).append(df['2012-12-01']).reset_index()
df = df.rename(columns={0: "Roubos"})
df
```
